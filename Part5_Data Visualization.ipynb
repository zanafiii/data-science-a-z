# Part 5 - Data Visualization

## Data Visualization with Matplotlib and Seaborn

### Read Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv("sales_records.csv")
df
```

```python
df.info()
```

```python
## Comparison - Bar Chart
```

```python
dfg = df.groupby("Sales Channel").agg({"Total Profit":"sum"}).reset_index()
dfg
```

```python
plt.bar(x = "Sales Channel", height = "Total Profit", data = dfg)
```

```python
sns.barplot(x = "Sales Channel", y = "Total Profit", data = dfg)
```

```python
## Comparison - Multiline Chart
```

```python
df.info()
```

```python
df["Order Month"] = pd.to_datetime(df["Order Date"]).dt.to_period('M').astype(str)
df.info()
```

```python
plt.figure(figsize = (12,6))
plt.xticks(rotation = 90)

sns.lineplot(x = "Order Month", y = "Total Profit", data = df, label = "Total Profit")
sns.lineplot(x = "Order Month", y = "Total Revenue", data = df, label = "Total Revenue")
plt.legend()
```
